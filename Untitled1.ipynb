```python
from scipy.stats import ttest_ind

# 
feature_columns = [col for col in pca_df.columns if col != "ChurnStatus"]

#  p 
t_test_results = {}

#  t  p 
for feature in feature_columns:
    stat, p_value = ttest_ind(
        pca_df[pca_df["ChurnStatus"] == 0][feature],
        pca_df[pca_df["ChurnStatus"] == 1][feature],
        equal_var=False  # 
    )
    t_test_results[feature] = p_value

#  p 
sorted_results = sorted(t_test_results.items(), key=lambda x: x[1])

# 
print("Feature-wise T-test P-values:")
for feature, p_value in sorted_results:
    print(f"{feature}: {p_value:.4f}")
```

```python
from scipy.stats import mannwhitneyu

# 
feature_columns = [col for col in pca_df.columns if col != "ChurnStatus"]

#  p 
mannwhitney_results = {}

#  Mann-Whitney U  p 
for feature in feature_columns:
    try:
        stat, p_value = mannwhitneyu(
            pca_df_resampled[pca_df_resampled["ChurnStatus"] == 0][feature],
            pca_df_resampled[pca_df_resampled["ChurnStatus"] == 1][feature]
        )
        mannwhitney_results[feature] = p_value
    except Exception as e:
        print(f"Error processing feature {feature}: {e}")

#  p 
sorted_results = sorted(mannwhitney_results.items(), key=lambda x: x[1])

# 
print("Feature-wise Mann-Whitney U-test P-values:")
for feature, p_value in sorted_results:
    print(f"{feature}: {p_value:.4f}")
```

```python
X_train = pca_df.drop('ChurnStatus', axis=1).values
y_train = pca_df['ChurnStatus'].values
X_test = pca_test.drop('ChurnStatus', axis=1).values
y_test = pca_test['ChurnStatus'].values
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold

stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
param_grid = {
    'max_depth': [5, 10, 15],
    'max_features': ['sqrt'],
    'min_samples_leaf': [5, 10],
    'min_samples_split': [5, 10],
    'n_estimators': [100, 200, 300],
    'class_weight': ['balanced_subsample']
}

rf = RandomForestClassifier(class_weight='balanced')
rf_grid = GridSearchCV(rf, param_grid, cv=stratified_kfold, scoring='roc_auc', verbose=0, n_jobs=-1)
rf_grid.fit(X_train, y_train)
rf_best = rf_grid.best_estimator_
```

```python
from sklearn.model_selection import learning_curve



train_sizes, train_scores, test_scores = learning_curve(
    rf_best, X_train, y_train, cv=stratified_kfold, scoring='accuracy', n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10)
)


train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)


plt.figure(figsize=(8, 6))
plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color="r")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve for Random Forest")
plt.legend(loc="best")
plt.grid()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay


y_pred = rf_best.predict(X_test)


cm = confusion_matrix(y_test, y_pred)


disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf_best.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, matthews_corrcoef, roc_auc_score
from sklearn.metrics import classification_report, roc_curve, auc
y_pred = rf_best.predict(X_test)
y_proba = rf_best.predict_proba(X_test)[:, 1]  


precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
mcc = matthews_corrcoef(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_proba)


print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print(f"MCC: {mcc:.2f}")
print(f"ROC AUC: {roc_auc:.2f}")

#  ROC 
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
param_grid = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [2, 3, 5],
    'learning_rate': [0.001, 0.01, 0.02],
    'subsample': [0.5, 0.6, 0.8],
    'colsample_bytree': [0.5, 0.6, 0.8],
    'reg_alpha': [0, 0.1, 1, 10],          #  L1 
    'reg_lambda': [1, 5, 10, 20],          #  L2 
    'min_child_weight': [1, 5, 10] 
}

xgb = XGBClassifier(eval_metric="logloss", random_state=42)

grid_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    n_iter=300,
    scoring='f1',
    cv=5,
    verbose=1,
    n_jobs=-1,
    random_state=42
)

grid_search.fit(X_train, y_train)


best_xgb = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)


y_pred = best_xgb.predict(X_test)
y_proba = best_xgb.predict_proba(X_test)[:, 1]


print("Classification Report:\n", classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
```
